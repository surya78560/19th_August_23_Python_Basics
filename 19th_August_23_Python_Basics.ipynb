```python
#This is the solution of Qno 1

x = 25
y = 5
def swap_variables (x,y):
    print(f"Before swap x is {x} and y is {y}")
    x,y = y,x
    print (f"After swap x is {x} and y is {y}") 

swap_variables(x,y)    
```

```python
'''QNo_3:- Write a Python program that converts 
temperatures from Celsius to Fahrenheit. 
Take the temperature in Celsius as input,
store it in a variable, convert it to Fahrenheit,
and display the result.'''
```

```python
#This is the solution of Qno3

Celsius_temperature = float(input("enter the temperature in celsius:-"))
Fahrenheit_temperature = ((Celsius_temperature *9/5)+35)
print(f"{Celsius_temperature} Celsius is equal to {Fahrenheit_temperature:.2f} Fahrenheit")
```

```python
''' QNo2:- 2. Create a program that takes a 
sentence from the user and counts the number of 
vowels (a, e, i, o, u) in the string.'''
```

```python
# This is the solution of QNo2.

sentence = input("enter a sentence:-")
vowel_count = 0
for char in sentence:
    character_lower = char.lower()
    if character_lower in "aeiou":
        vowel_count += 1

print(f"Numbers of Vowels :- {vowel_count}")
```

```python
''' QNo_3:- Given a string, reverse the order 
of characters using string slicing 
and print the reversed string.'''
```

```python
#This is the solution of QNo_3.

input_string = str(input("enter a string:-"))
reversed_string = input_string [::-1]
print(f"Reversed string is :- {reversed_string}")
```

```python
''' QNo_4:- Write a program that takes a 
string as input and checks if it is a 
palindrome (reads the same forwards and backwards).'''
```

```python
# This is the solution of QNo_4.

input_string = str(input("Enter a string:-"))
cleaned_string = input_string.replace(" ","").lower()
reversed_string = cleaned_string[::-1]
if cleaned_string == reversed_string:
    print("The input string is palindrome.")
    
else:
    print("The input string is not a palindrome.")
```

```python
''' QNo_5 :- Create a program that takes a string
as input and removes all the spaces from it.
Print the modified string without spaces.'''
```

```python
# This is the solution of QNo_5.

input_string = str(input("enter a string :-"))
modified_string = ""
for char in input_string:
    if char != " ":
        modified_string += char

print(f"Modified string without spaces :- {modified_string}")
```
